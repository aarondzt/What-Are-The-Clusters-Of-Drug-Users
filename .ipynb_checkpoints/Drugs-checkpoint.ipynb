## Clustering on demography, personality, and drug use: What patterns are associated with hard drug use (e.g. heroin, crack, etc.)?

### Abstract
Understanding the personality traits that are predictive of illegal drug usage is important for public health and guidence of policy-making. In Fehrman et al., the authors collect survey data on demographic factors, personality traits, and drug usage, and then run several machine learning models to analyze the traits that are predictive of the usage of individual and clusters of drugs. In a previous project, I attempted to classify users vs. non-users of illegal drugs based on demography and personality. Here, I attempt to use PCA and clustering to examine the inherent structure of the data, with an emphasis on finding patterns associated with hard drug use (e.g. use of drugs such as heroin and crack that are known to be associated with negative life outcomes). I used PCA to reduce the dimensionality of the personality and drug use features. The first PC of personality was associated with conscientiousness, and the second with introversion. The first PC of drug use differentiated users from non-users, and the second differentiated users of hard drugs with users of psychdelic drugs (e.g. ecstasy, mushrooms, LSD). Then, I used K-means clustering to identify meaningful and informative clusters. Using the elbow method with distortion as a metric to guide my clustering decisions, I settled on k = 3, as this divided the observations into three groups: (a) non-users, (b) hard drug users, and (c) psychdelic drug users. I visualized these clusters against educational attainment and conscientiousness and found that hard drug users exhibit lower education and conscientiousness, psychedelic drug users exhibit average education and conscientiousness, and non-users exhibit higher education and conscientiousness. As far as public policy goes, this analysis suggests that increasing the quality of education and student retention might be an effective solution to deter the use of hard drugs (this would probably also raise conscientiousness). However, this would have to be proven with causal experiments.

### Introduction
The authors of the paper collected survey data on five demographic factors (age, gender, education, country, ethnicity) and seven personality traits (neuroticism, extraversion, openness to experience, agreeableness, conscientiousness from the [Five Factor Model](https://www.annualreviews.org/doi/abs/10.1146/annurev.ps.41.020190.002221), and impulsivity and sensation seeking from two other personlity inventories). The Five Factor model is a well-validated metric of personality that has shown to be predictive of real-world outcomes in many studies. The impulsivity and sensation seeking metrics are specifically tailored to capture traits known to be associated with drug use. In addition to this, the authors collected data on the usage of 16 psychoactive substances (alcohol, amphetamines, amyl nitrite, benzodiazepines, caffeine, cannabis, cocaine, crack, ecstasy, heroin, ketamine, legal highs, LSD, methadone, mushrooms, nicotine, volatile substances), plus two controls (chocolate and a fictitious substance called semer). In the dataset provided by the [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Drug+consumption+%28quantified%29), each feature was quantified and mean-centered. Some features were ordinal such as age (from bracket "18-24" to "65+"), education (from 
"Left school before 16 years" to "Doctorate"), and drug use (one of seven responses from "Never used" to "Used in the last day." Other features, such as country and ethnicity, were quantified, although the quantification is neccessarily arbitrary.

### Methods 1.1 Data loading

```python
# Read csv
import pandas as pd
csv = pd.read_csv('Data/Drugs.csv', header = None)
print(csv.shape)
```

```python
# Name columns
columns = ['Age', 'Female', 'Education', 'Country', 'Ethnicity', \
           'Neuroticism', 'Extraversion', 'Openness', 'Agreeableness', 'Conscientiousness', \
           'Impulsivity', 'SensationSeeking', \
           'Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine', 'Cannabis', 'Chocolate', \
           'Cocaine', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', \
           'Nicotine', 'Semer', 'VolatileSubstances']
```

```python
# Create DataFrame
data = pd.DataFrame(data = csv)
data.columns = columns
```

### Methods 1.2 Data cleaning and scaling

```python
# Drop multi-category, non-ordinal features
data = data.drop(['Country', 'Ethnicity'], axis = 1)
```

```python
# Check value count for Semer, a fictitious drug included to weed out participants responding improperly
print(data.Semer.value_counts())
print(data.shape)
# Remove these rows
data = data[data['Semer'] == 0]
print(data.shape)
# Drop semer column
data = data.drop('Semer', axis = 1)
data = data.reset_index(drop = True)
print(data.shape)
```

```python
# Name columns
columns = ['Age', 'Female', 'Education', \
           'Neuroticism', 'Extraversion', 'Openness', 'Agreeableness', 'Conscientiousness', \
           'Impulsivity', 'SensationSeeking', \
           'Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine', 'Cannabis', 'Chocolate', \
           'Cocaine', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', \
           'Nicotine', 'VolatileSubstances']
print(len(columns))
```

```python
# Scale data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_raw = data
data = scaler.fit_transform(data)
data = pd.DataFrame(data)
data.columns = columns
```

### Methods 1.3 PCA

```python
# Principal component analysis on personality
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data[['Neuroticism', 'Extraversion', 'Openness', 'Agreeableness', 'Conscientiousness', \
           'Impulsivity', 'SensationSeeking']].T)
pca_df = pd.DataFrame(pca.components_)
pca_df = pca_df.T
```

```python
# Plot cumulative explained variance per component
import numpy as np
from matplotlib import pyplot as plt
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance');
```

```python
# Add PCs to main data
data['PersonalityPC1'] = pca_df[0]
data['PersonalityPC2'] = pca_df[1]
```

```python
# PC1 = Conscientious
data[['Neuroticism', 'Extraversion', 'Openness', 'Agreeableness', 'Conscientiousness', \
           'Impulsivity', 'SensationSeeking', 'PersonalityPC1']].corr()['PersonalityPC1'].sort_values().round(2)
```

```python
# PC2 = Introversion
data[['Neuroticism', 'Extraversion', 'Openness', 'Agreeableness', 'Conscientiousness', \
           'Impulsivity', 'SensationSeeking', 'PersonalityPC2']].corr()['PersonalityPC2'].sort_values().round(2)
```

```python
# Principal component analysis on drugs
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data[['Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine', 'Cannabis', 'Chocolate', \
           'Cocaine', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', \
           'Nicotine', 'VolatileSubstances']].T)
pca_df = pd.DataFrame(pca.components_)
pca_df = pca_df.T
pca_df.head()
```

```python
# Plot cumulative explained variance per component
import numpy as np
from matplotlib import pyplot as plt
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance');
```

```python
# Add PCs to main data
data['DrugsPC1'] = pca_df[0]
data['DrugsPC2'] = pca_df[1]
```

```python
# PC1 = Non drug users vs. drug users
data[['Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine', 'Cannabis', 'Chocolate', \
           'Cocaine', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', \
           'Nicotine', 'VolatileSubstances', 'DrugsPC1']].corr()['DrugsPC1'].sort_values().round(2)
```

```python
# PC2 = Hard drug users vs. psychedelic drug users
data[['Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine', 'Cannabis', 'Chocolate', \
           'Cocaine', 'Crack', 'Ecstasy', 'Heroin', 'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', \
           'Nicotine', 'VolatileSubstances', 'DrugsPC2']].corr()['DrugsPC2'].sort_values().round(2)
```

```python
# Rename PCs based on interpretations
columns = ['Age', 'Female', 'Education', 'Neuroticism', 'Extraversion', 'Openness',
       'Agreeableness', 'Conscientiousness', 'Impulsivity', 'SensationSeeking',
       'Alcohol', 'Amphetamines', 'AmylNitrite', 'Benzodiazepines', 'Caffeine',
       'Cannabis', 'Chocolate', 'Cocaine', 'Crack', 'Ecstasy', 'Heroin',
       'Ketamine', 'LegalHighs', 'LSD', 'Methadone', 'Mushrooms', 'Nicotine',
       'VolatileSubstances', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser']
data.columns = columns
```

```python
# Scale again after addition of PCs
scaler = StandardScaler()
data = scaler.fit_transform(data)
data = pd.DataFrame(data)
data.columns = columns
```

### Methods 1.4 Clustering

```python
# Find optimal number of clusters for K-Means
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
from warnings import simplefilter
simplefilter(action = 'ignore', category = FutureWarning)

model = KMeans(random_state = 20201103)
visualizer = KElbowVisualizer(model, k = (2,10), timings = False)
visualizer.fit(data[['Age', 'Female', 'Education', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser']])
elbow_visualizer = visualizer.show()
```

```python
# K-means clustering on demographic features + personality/drug PC features
k_means = KMeans(n_clusters = 3, random_state = 20201103)
k_means.fit(data[['Age', 'Female', 'Education', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser']])
clusters = k_means.predict(data[['Age', 'Female', 'Education', 
                                 'PC_Conscientious', 'PC_Introverted', 
                                 'PC_NonDrugUserVsDrugUser', 'PC_HardDrugUserVsPsychDrugUser']])
data['KMeans_Cluster'] = clusters
```

```python
# Group by cluster
data_kmeans = data[['Age', 'Female', 'Education', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser', 'KMeans_Cluster']]
data_kmeans.groupby('KMeans_Cluster').mean().round(2)
```

Cluster 0 = Psychedelic drug users

Cluster 1 = Non drug users

Cluster 2 = Hard drug users

```python
# Choosing epsilon for DBSCAN (https://towardsdatascience.com/machine-learning-clustering-dbscan-determine-the-optimal-value-for-epsilon-eps-python-example-3100091cfbc)
from sklearn.neighbors import NearestNeighbors
neighbors = NearestNeighbors(n_neighbors = 2)
neighbors = neighbors.fit(data.drop('KMeans_Cluster', axis = 1))
distances, indices = neighbors.kneighbors(data.drop('KMeans_Cluster', axis = 1))
distances = np.sort(distances, axis = 0)
distances = distances[:,1]
plt.plot(distances)
```

```python
# DBSCAN clustering on demographic features + personality/drug PC features, eps = 2 = 3 clusters like K-Means
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 2)
dbscan.fit(data[['Age', 'Female', 'Education', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser']])
np.unique(dbscan.labels_)
```

```python
# Group by cluster
data['DBSCAN_Cluster'] = dbscan.labels_
data_dbscan = data[['Age', 'Female', 'Education', 'PC_Conscientious', 'PC_Introverted', 'PC_NonDrugUserVsDrugUser',
       'PC_HardDrugUserVsPsychDrugUser', 'DBSCAN_Cluster']]
data_dbscan.groupby('DBSCAN_Cluster').mean().round(2)
```

DBSCAN does not differentiate between hard drug and psychedelic drug users as dramatically as K-Means

DBSCAN also produces very imbalanced clusters, with one cluster of n = 4 of very old hard drug users

```python
# Rename KMeans_Clusters
pd.set_option('mode.chained_assignment', None)
data_kmeans.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].astype(str)
data_kmeans.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('0','Psychedelic drug users')
data_kmeans.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('1','Non drug users')
data_kmeans.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('2','Hard drug users')
data.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].astype(str)
data.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('0','Psychedelic drug users')
data.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('1','Non drug users')
data.loc[:, 'KMeans_Cluster'] = data_kmeans.loc[:, 'KMeans_Cluster'].str.replace('2','Hard drug users')
```

```python
# Sanity check means (note: scaling means mean = std from sample mean)
print('Hard drug users heroin use', round(data[data['KMeans_Cluster'] == 'Hard drug users']['Heroin'].mean(),1), 'std compared to mean')
print('Psychedelic drug users heroin use', round(data[data['KMeans_Cluster'] == 'Psychedelic drug users']['Heroin'].mean(),1), 'std compared to mean')
print('Non drug users heroin use', round(data[data['KMeans_Cluster'] == 'Non drug users']['Heroin'].mean(),1), 'std compared to mean')
print('Hard drug users ecstasy use', round(data[data['KMeans_Cluster'] == 'Hard drug users']['Ecstasy'].mean(),1), 'std compared to mean')
print('Psychedelic drug users ecstasy use', round(data[data['KMeans_Cluster'] == 'Psychedelic drug users']['Ecstasy'].mean(),1), 'std compared to mean')
print('Non drug users ecstasy use', round(data[data['KMeans_Cluster'] == 'Non drug users']['Ecstasy'].mean(),1), 'std compared to mean')
```

```python
# Sanity check sums (from original unscaled drug values referring to frequency of use, 0 = never)
data_raw['KMeans_Cluster'] = data['KMeans_Cluster']
print('Hard drug users heroin use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Hard drug users']['Heroin'].sum(),1))
print('Psychedelic drug users heroin use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Psychedelic drug users']['Heroin'].sum(),1))
print('Non drug users heroin use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Non drug users']['Heroin'].sum(),1))
print('Hard drug users ecstasy use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Hard drug users']['Ecstasy'].sum(),1))
print('Psychedelic drug users ecstasy use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Psychedelic drug users']['Ecstasy'].sum(),1))
print('Non drug users ecstasy use sum is', round(data_raw[data_raw['KMeans_Cluster'] == 'Non drug users']['Ecstasy'].sum(),1))
```

### Conclusions
The above figure showcases the effectiveness of K-Means clustering in identifying patterns in the data. Rather than 16 individual drugs, the figure is much more illustrative when participants are clustered into the groups of hard drug users, psychedelic drug users, and non drug users.

Participants with lower conscientiousness and education levels were more likely to use hard drugs, and use them more frequently. In the middle levels of conscientiousness and education were participants who stayed away from hard drugs such as heroin and crack, but did experiment with psychedelic drugs such as ecstasy, mushrooms, LSD, and cannabis. Finally, participants with higher levels of conscientiousness and education were the most likely to be non-users.

### Limitations
* In interpretation, the PCs referring to "conscientiousness" and "introversion" must be thought of as referring to their associated traits, such as lack of impulsivity and neuroticism, respectively.
* Similarly, the PCs referring to drug users do not entail that, for example, a high value on PC_NonDrugUserVsDrugUser for a participant means that they have never used a drug, just that it is very unlikely that they have. This is why after clustering, there is some "misclassification," e.g. "Non drug users heroin use sum is 58."
* Effects of age and gender were not controlled for, so some of these patterns may be explained by multicollinear correlations with age and gender.

### Future Directions
In the future, I would like to regress out the effects of age and gender (two factors you cannot intervene on, as opposed to say, increasing education funding) on the drug use PCs in order to isolate the effects of education and personality and mitigate the distortionary effects of these variables (e.g. age is associated with being a non user, but that could just be due to the fact that people use fewer drugs as they age so they are less likely to respond that they have done a drug in the past month).
